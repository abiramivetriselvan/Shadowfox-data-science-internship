```python
import matplotlib.pyplot as plt
colors=['Red','Yellow','Blue','Green']
count=[1000,500,700,1100]
plt.bar(colors,count)
plt.title('Bar Plot')
plt.xlabel('Colors')
plt.ylabel('Count')
plt.show()
```

Swarm Plot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
categories = ['A', 'B', 'C', 'D']
data = {
    'category': np.random.choice(categories, size=100),
    'value': np.random.rand(100) * 100
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create and display a swarm plot
sns.swarmplot(x='category', y='value', data=df)
plt.title('Swarm Plot of Synthetic Data')
plt.show()
```
