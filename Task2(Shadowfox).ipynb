```python
data['date']=pd.to_datetime(data['date'])
```

```python
num_cols=data.select_dtypes(include=['number']).sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.title('Distrubution of the polutants')
plt.pie(num_cols,labels=num_cols.index,autopct='%1.1f%%')
```

```python
num_cols2=['co','no','no2','o3','so2','pm2_5','pm10','nh3']
for col in num_cols2:
  plt.hist(data[col],bins=20,edgecolor='black')
  plt.xlabel(col)
  plt.ylabel('Frequency')
  plt.title('Frequency of {col}')
  plt.show()
```

```python
import seaborn as sns

for col in num_cols2:
  sns.boxplot(data=data,y=data[col])
  plt.title(f'Boxplot for {col}')
  plt.show()
```

```python
for i in range (len(num_cols2)):
  for j in range(i+1,len(num_cols2)):
    plt.scatter(data=data,x=num_cols2[i],y=num_cols2[j],edgecolors='black',color='cyan')
    plt.xlabel(num_cols[i])
    plt.ylabel(num_cols[j])
    plt.title(f'Scatter plot of {num_cols2[i]} vs {num_cols2[j]}')
    plt.show()
```

```python
correlation_matrix=data[num_cols2].corr()
sns.heatmap(correlation_matrix,annot=True,fmt=".2f",cmap='coolwarm',annot_kws={"size": 10})
plt.title('Correlation Matrix of Pollutants')
plt.show()
```

```python
#Predicting AQI Categories: Based on predicted pollutant levels
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def get_aqi_category(pm2_5, pm10, o3, no2, so2, co, nh3):
    if pm2_5 < 50:
        return "Good"
    elif pm2_5 < 100:
        return "Moderate"
    elif pm2_5 < 150:
        return "Unhealthy for Sensitive Groups"
    elif pm2_5 < 200:
        return "Unhealthy"
    elif pm2_5 < 300:
        return "Very Unhealthy"
    else:
        return "Hazardous"

data['AQI_Category'] = data.apply(lambda row: get_aqi_category(row['pm2_5'], row['pm10'], row['o3'], row['no2'], row['so2'], row['co'], row['nh3']), axis=1)
print(data[['date', 'pm2_5', 'pm10', 'o3', 'no2', 'so2', 'co', 'nh3', 'AQI_Category']].head(10))  # Showing 10 rows

plt.figure(figsize=(15, 5))
sns.lineplot(x='date', y='pm2_5', data=data, label='PM2.5', marker='o', markersize=4)
sns.lineplot(x='date', y='pm10', data=data, label='PM10', marker='o', markersize=4)
sns.lineplot(x='date', y='o3', data=data, label='O3', marker='o', markersize=4)
sns.lineplot(x='date', y='no2', data=data, label='NO2', marker='o', markersize=4)
sns.lineplot(x='date', y='so2', data=data, label='SO2', marker='o', markersize=4)
sns.lineplot(x='date', y='co', data=data, label='CO', marker='o', markersize=4)
sns.lineplot(x='date', y='nh3', data=data, label='NH3', marker='o', markersize=4)



plt.xlabel('Date')
plt.ylabel('Pollutant Levels')
plt.title('Timeline of Pollutant Levels')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
